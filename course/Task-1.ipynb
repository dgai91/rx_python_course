Fama French¬∂

ùëü=ùëÖùëì+ùëèùëö‚ãÖMER+ùëèùë†‚ãÖSMB+ùëèùë£‚ãÖHML+ùëèùëù‚ãÖRMW+ùëèùëñ‚ãÖCMA

ùëü  is expected portfolio return
ùëèùëö ,  ùëèùë† ,  ùëèùë£ ,  ùëèùëù ,  ùëèùëñ  are factor sensitivities

Factors Analysis
MER  is excess return of a broad market over the risk-free rate ( ùëÖùëö‚àíùëÖùëì )(market factor)

SMB  is the average return on the small stock portfolios minus big stock portfolios (size factor)

HML  is the average return on high book value portfolios minus low book value (value factor)

RMW  is the average return on robust minus weak operating profitability portfolios (profitability factor)

CMA  Conservative minus aggresive investments, proxied by the y/y growth in company asset value (investment factor)

```python
#read the stock data
ff_factors = pdr.DataReader('F-F_Research_Data_5_Factors_2x3','famafrench', start_date, end_date)[0]
ff_factors
```

```python
stockret_mon.index = ff_factors.index
AAPL_stockret_mon.index = ff_factors.index
```

```python
stockret_mon
```

```python
BABA_data_on_ff =pda.merge(stockret_mon, ff_factors, on='Date')
```

```python
BABA_data_on_ff
```

```python
BABA_data_on_ff[['Mkt-RF','SMB','HML','RMW','CMA','RF']] = BABA_data_on_ff[['Mkt-RF','SMB','HML','RMW','CMA','RF']]/100
```

```python
BABA_data_on_ff
```

```python
X=BABA_data_on_ff[['Mkt-RF','SMB','HML','RMW','CMA']]
y=BABA_data_on_ff['Adj Close']

X_sm = smd.add_constant(X)
```

```python
model = smd.OLS(y,X_sm)
results = model.fit()
results.summary()
```
