```python
import pdfplumber
from gensim import corpora, models
import spacy
```

## read pdfs

```python
pdf1 = pdfplumber.open('../../data/text.pdf')
pdf2 = pdfplumber.open('../../data/jin2019.pdf')
```

## load an EN version scapy model

```python
nlp = spacy.load('en_core_web_sm')
```

## generate text lists for all tasks, do NER task using spacy model.

```python
all_texts1, all_entities1 = [], []
for pid in range(len(pdf1.pages)):
    text = pdf1.pages[pid].extract_text()
    text = text.replace('-\n', '').replace('\n', '')
    doc = nlp(text)
    all_texts1.extend([word.lemma_ for word in doc])
    for x in doc.ents:
        if (x.text, x.label_) not in all_entities1:
            if x.label_ != 'CARDINAL':
                all_entities1.append((x.text, x.label_))
for line in all_entities1:
    print(line)
```

```python
all_texts2, all_entities2 = [], []
for pid in range(len(pdf2.pages)):
    text = pdf2.pages[pid].extract_text()
    text = text.replace('-\n', '').replace('\n', '')
    doc = nlp(text)
    all_texts2.extend([word.lemma_ for word in doc])
    for x in doc.ents:
        if (x.text, x.label_) not in all_entities2:
            if x.label_ != 'CARDINAL':
                all_entities2.append((x.text, x.label_))
for line in all_entities2:
    print(line)
```

## create a dict using all texts

```python
dictionary = corpora.Dictionary([all_texts1, all_texts2])
```

## using gensim generate a set of BoW models, sorted by word count 

```python
corpus1 = dictionary.doc2bow(all_texts1)
corpus1 = sorted(corpus1, key=lambda x: x[1], reverse=True)
for word in corpus1:
    print(dictionary[word[0]], word[1])
```

```python
corpus2 = dictionary.doc2bow(all_texts2)
corpus2 = sorted(corpus2, key=lambda x: x[1], reverse=True)
for word in corpus2:
    print(dictionary[word[0]], word[1])
```

## generate a tfidf model using 2 corpus

```python
tfidf = models.TfidfModel([corpus1, corpus2])
```

## compute tfidf

```python
corpus1_tfidf = sorted(tfidf[corpus1], key=lambda x: x[1], reverse=True)
for word in corpus1_tfidf:
    print(dictionary[word[0]], word[1])
```

```python
corpus2_tfidf = sorted(tfidf[corpus2], key=lambda x: x[1], reverse=True)
for word in corpus2_tfidf:
    print(dictionary[word[0]], word[1])
```
